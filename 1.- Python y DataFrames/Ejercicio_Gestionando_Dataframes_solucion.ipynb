## Como has visto anteriormente, las variables Gender, ST y Education tienen valores "raros", crea nuevas variables con esos valores corregidos

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html




```python
df.loc[df['GENDER'].isin(['M','Male','MA']), 'GENDER'] = 'M'
df.loc[df['GENDER'].isin(['Femal','female','F']), 'GENDER'] = 'F'
df['GENDER'].value_counts()
```

```python
df.loc[df['ST'].isin(['AZ']), 'ST'] = 'Arizona'
df.loc[df['ST'].isin(['Cali']), 'ST'] = 'California'
df.loc[df['ST'].isin(['WA']), 'ST'] = 'Washington'
df['ST'].value_counts()
```

```python
df.loc[df['Education'] == 'Bachelors', 'Education'] = 'Bachelor'
df['Education'].value_counts()
```

## INDICES

N√∫mero de serie de cada l√≠nea de un dataframe. Originalmente, cuando importas o creas una nueva tabla, es una columna que va desde 0 hasta la longitud (n√∫mero de l√≠neas) del propio dataframe.

En ocasiones nos interesa que una de nuestras columnas sea el √≠ndice, para ello existe el m√©todo .set_index('columna'), que nos permite cambiar el √≠ndice de la tabla.

A trav√©s del m√©todo .reset_index() volvemos la acci√≥n para atr√°s
